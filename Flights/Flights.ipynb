```python
with open("data_f.json", "r") as mf:
    data = json.loads(mf.read())
    
df = pd.DataFrame(data["states"])
```

```python
df.columns = ["icao24","callsign", "origin_country","time_position", "last_contact","longitude", "latitude", "baro_altitude", "on_ground", "velocity", "true_track", "vertical_rate", "sensors", "geo_altitude", "squawk", "spi", "position_source"   ]
```

```python
df.head()
```

```python
m = folium.Map(location=[df["longitude"], df["latitude"]])
m
m.save('index.html')
```

```python
app = Flask("_name_")
@app.route('/')
def index():
    start_coords = (46, 142)
    folium_map =  folium.Map(location=start_coords, zoom_start = 14)
    return folium_map._repr_html_()
```
